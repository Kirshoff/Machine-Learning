### Multiple Regression and Prediction


```python
import pandas as pd
from sklearn.linear_model import LinearRegression as lr
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
df=pd.DataFrame(iris)
df.head()
```

```python
sns.pairplot(df, hue="species");
```

```python
df.species.unique()
```

```python
df.species.replace(to_replace=['setosa', 'versicolor', 'virginica'], value=[1, 2, 3],inplace=True)
```

```python
df.head()
```

```python
X = df[['sepal_length', 'sepal_width','petal_length','petal_width']]
y = df['species']
regr = lr()
regr.fit(X, y)
```

```python
species={1:'setosa',2:'versicolor',3:'virginica'}
```

```python
# We will predict the type of leaf according to the leaf sizes.
# input_data= [sepal_length,sepal_width,petal_length,petal_width]
    
input_data=[6.3, 3.6, 5.8, 2.5]
```

```python
#predict the spice of the plant
predicted_spices = regr.predict([input_data])
spice = species[round(predicted_spices[0])]
print(f"Its spice is '{spice}' with %{round(abs(predicted_spices[0])-round(predicted_spices[0]),3)*100} accuracy.")
```
