## Evaluation Metrics

```python
def eval_metrics(actual, pred):
    rmse = np.sqrt(mean_squared_error(actual, pred))
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    r2score = r2_score(actual, pred)
    print(f"R2 Score: {r2score}",
          f"Mean Absolute Error(MAE): {mae}",
          f"Mean Square Error(MSE): {mse}",
          f"Root Mean Square Error(RMSE): {rmse}",sep="\n")
```

```python
eval_metrics(y_test, y_pred)
```

```python
eval_metrics(y_train, y_pred_train)
```


## 2.Ridge Regression
 Varyansi azaltmak icin modelin biasini artirmak amactir
        Pros
        
    * Trades variance for bias (i.e. in presence of co-linearity, it is worth to have biased results, in order to lower the variance.)
    
    * Prevents over fitting
    
        Cons
        
    * Increases bias
    
    * Need to select perfect alpha (hyper parameter)
    
    * Model interpret-ability is low

## 3.Lasso Regression

    * n:size, p:diamension
    
           Pros
           
    *Select features, by shrinking co-efficient towards zero.
    
    *Avoids over fitting
    
            Cons
            
    *Selected features will be highly biased.
    
    *For n<<p (n-number of data points, p-number of features), LASSO selects at most n features.
    
    *LASSO will select only one feature from a group of correlated features, the selection is arbitrary in nature.
    
    *For different boot strapped data, the feature selected can be very different.
    
    *Prediction performance is worse than Ridge regression.

```python
from sklearn.linear_model import Lasso, LassoCV
```

```python
lasso_model = Lasso()
lasso_model.fit(X_train, y_train)
y_pred = lasso_model.predict(X_test)
eval_metrics(y_test, y_pred)
```

```python
accuraries = cross_val_score(estimator=lasso_model, X=X_train, y=y_train, cv=10)
accuraries.mean()
```

## 4.Elastic Net

    * Residual MSE + α⋅Ridge Penalty + (1−α)⋅LASSO Penalty
    
    * if α = 0, it works like LASSO. if α = 1, it works like Ridge. It is a hybrid model of Ridge&LASSO.
    
    * Pros: Doesn’t have the problem of selecting more than n predictors when n<<p, whereas LASSO saturates when n<<p.
    
    * Cons: Computationally more expensive than LASSO or Ridge.

```python
from sklearn.linear_model import ElasticNetCV
elastic_model = ElasticNetCV(alphas=alpha_space, l1_ratio=[.1, .5, .7,.9, .95, .99, 1])
```

```python
elastic_model.fit(X_train,y_train)
```

```python
elastic_model.l1_ratio_
```

```python
elastic_model.alpha_
```

```python
y_pred = elastic_model.predict(X_test)
```

```python
eval_metrics(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
accuraries = cross_val_score(estimator=elastic_model, X=X_train, y=y_train, cv=10)
accuraries.mean()
```

### 5.Polynomial Regression

Residuals lara bakarak bu modele karar veririz

    * The presence of one or two outliers in the data can seriously affect the results of the nonlinear analysis.

    * These are too sensitive to the outliers.
    
    * Daha cok kucuk data setleri icin uygun bir modeldir.

                >> 2nd Degree Polynomial

```python
# we will use not scaled X (original X)
from sklearn.preprocessing import PolynomialFeatures
polynomial_converter = PolynomialFeatures(degree=2)
poly_features = polynomial_converter.fit_transform(X)
poly_features.shape
```

```python
X.shape
```

        >> 3 features var. Bunlari kendi arasinda birbiriyle carparak, x0 ile birlikte yeni 7 features(columns) daha ekledi.
        >> X.columns:(x1, x2, x3) ---> poly_features:(x0, x1, x2, x3, x1.x1, x2.x2, x3.x3, x1.x2, x1.x3, x2.x3)

```python
X.iloc[0]
```

```python
poly_features[0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
eval_metrics(y_test, y_pred)
```

```python
accuraries = cross_val_score(estimator=model, X=X_train, y=y_train, cv=10)
accuraries.mean()
```
